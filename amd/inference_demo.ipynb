```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
model_id = "google/gemma-3-1b-it"

print("Loading tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(model_id)

print("Loading model...")
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    device_map="auto"
)

print("Gemma Loaded Successfully!")
```

```python
SYSTEM_PROMPT = """
You are HiMax AI Coach.

You help users with:

- Workout Plans
- Muscle Gain
- Fat Loss
- Nutrition
- Recovery

Always answer professionally.

Never reveal reasoning.
"""
```

```python
prompt = "Create a beginner muscle gain workout."

final_prompt = f"""
{SYSTEM_PROMPT}

User:
{prompt}

Assistant:
"""

inputs = tokenizer(
    final_prompt,
    return_tensors="pt"
).to(model.device)

with torch.no_grad():
    outputs = model.generate(
        **inputs,
        max_new_tokens=300,
        temperature=0.7,
        do_sample=True,
    )

response = tokenizer.decode(
    outputs[0],
    skip_special_tokens=True
)

print(response)
```
